```python
df['Title_Abstract'] = df['Title'] + " " + df['Brief title'] + " " + df['Abstract']
```

```python
import pandas as pd
import string

# 2. Define a regular expression pattern for all punctuation
pattern = '[{}]'.format(string.punctuation)

# 3. Remove punctuation from column 'A'
df['Title_Abstract_punc'] = df['Title_Abstract'].str.replace(pattern, ' ', regex=True)

# 3. Remove punctuation from column 'A'
df['Title_Abstract_punc_lower'] = df['Title_Abstract_punc'].str.lower()

# 3. Remove punctuation from column 'A'
df['Title_Abstract_lower'] = df['Title_Abstract'].str.lower()

# 3. Remove punctuation from column 'A'
df['Title_Abstract_upper'] = df['Title_Abstract'].str.upper()

# 3. Remove punctuation from column 'A'
df['Title_Abstract_punc_upper'] = df['Title_Abstract_punc'].str.upper()


df['Title_Abstract_concat'] = df['Title_Abstract'] + " " + df['Title_Abstract_punc'] + " " + df['Title_Abstract_lower']  + " " + df['Title_Abstract_upper']  + " " + df['Title_Abstract_punc_upper']  + " " + df['Title_Abstract_punc_lower']

# 4. Display the updated DataFrame
df
```

```python
import spacy
import scispacy

# Load the scispaCy model
nlp = spacy.load("en_ner_bc5cdr_md")

def extract_diseases(text):
    doc = nlp(text)
    diseases = [ent.text for ent in doc.ents if ent.label_ == "DISEASE"]
    return diseases

df['Diseases'] = df['Title_Abstract_concat'].apply(extract_diseases)
```
